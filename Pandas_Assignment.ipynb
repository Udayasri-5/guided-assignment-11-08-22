```python
stock_df = pd.read_csv('/content/stocks.csv')
stock_df
```

```python
import matplotlib.pyplot as plt


stock_df.plot(x = 'Date', y = 'sp500', label = 'APPLE Stock Price', linewidth = 3);
plt.ylabel('Price')
plt.title('My first plotting exercise!')
plt.legend(loc="upper left")
plt.plot(stock_df['sp500'], 'r--');
plt.grid()
```

```python
stock_df.plot(x = 'Date', y =['AAPL','sp500','GOOG'], label = ['APPLE' ,'Stock ','Price'], linewidth = 3);
plt.ylabel('Price')
plt.title('My first plotting exercise!')
plt.legend(loc="upper left")
plt.plot(stock_df['sp500'], 'r--');
plt.grid()

plt.legend(ncol=4, loc='center', bbox_to_anchor=(0.5, 1.06))
plt.show()
```
